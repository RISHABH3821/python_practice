```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([1,2,3],[4,2,4])
plt.show()
```

```python
x = [1,2,3,4,5]
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("Message on x")
plt.ylabel("Message on Y")
plt.title("My title")
plt.grid()
plt.show()
```

```python
x = np.linspace(1,50,500)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("Message on x")
plt.ylabel("Message on Y")
plt.title("My title")
plt.grid()
plt.show()
```

```python
x = [1,2,3,4,5]
y=np.sin(x)
xnumber=np.linspace(0,7,15)
ynumber=np.linspace(-1,1,11)
plt.plot(x,y,color='r')
plt.xlabel("Message on x")
plt.ylabel("Message on y")
plt.title("My title")
plt.xticks(xnumber)
plt.yticks(ynumber)
plt.grid()
plt.show()
```

```python
x = np.linspace(0,2*np.pi,100)
y=np.sin(x)
xnumber=np.linspace(0,7,15)
ynumber=np.linspace(-1,1,11)
plt.plot(x,y,color='r')
plt.xlabel("Message on x")
plt.ylabel("Message on y")
plt.title("My title")
plt.xticks(xnumber)
plt.yticks(ynumber)
plt.grid()
plt.show()
```
