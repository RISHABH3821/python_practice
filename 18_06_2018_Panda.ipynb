weatherdata is a dictionary with multiple data[list] on one key.

```python
weatherdata={
    'day':['1/1/2018','2/12018'],
    'mintemp':[9,10],
    'maxtemp':[20,21],
    'windspeed':[5,6],
    'event':['rain','sunny']
}
df = pd.DataFrame(weatherdata)
print(df)
```
